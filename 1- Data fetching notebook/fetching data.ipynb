```python
import numpy as np
import pandas as pd
import json, sys, regex, os
import random
import requests
```

```python
url = 'https://recruitment.aimtechnologies.co/ai-tasks'
```

```python
original_data = pd.read_csv('/content/dialect_dataset.csv')
original_data.head()
```

```python
original_data.shape
```

**Requesting data as batches, each batch is 1000 tweet.**

```python
%%time

size = len(original_data)
tweets = []
start = 0
# Iterate over all data with batches, each one = 1000 row data.
for b, end in enumerate(range(1000,size, 1000)):
    batch = original_data['id'][start:end].tolist()
    batch = [str(num) for num in batch]
    response_API = requests.post(url, json= batch)
    print("Batch # {:<8}   start {:<8} end  {:<8} is :{:<8}".format(b+1, start, end, response_API.status_code))
    data = response_API.content
    parse_json = json.loads(data)
    tmp = [parse_json[id] for id in batch if id in parse_json]

    tweets.extend(tmp)
    start = end

# we stopped at the end variable which is indicate the last request number dat.So we will take the rest data starting form end.
batch = original_data['id'][end:].tolist()
batch = [str(num) for num in batch]
response_API = requests.post(url, json= batch)
data = response_API.content
parse_json = json.loads(data)
tmp = [parse_json[id] for id in batch if id in parse_json]
tweets.extend(tmp)
print("Batch # {:<8}   start {:<8} end  {:<8} is :{:<8}".format(b+2, start, size, response_API.status_code))
```

```python
# save data and upload on my drive
data = pd.DataFrame({'id': original_data['id'].tolist(), 
                     'tweets': tweets,
                     'dialect': original_data['dialect'].tolist()})

data.to_csv('Ara_dialect.csv', index=False)
```
